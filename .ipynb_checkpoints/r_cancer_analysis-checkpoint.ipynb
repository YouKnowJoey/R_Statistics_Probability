## Analyze Cancer Data

```r
# Imported library for data analysis
library("tidyverse")
```

```r
library("ggformula")
```

```r
cancer_data
```

```r
cancer_data %>% select(organ)
```

```r
# Summarize statistics for a numeric column using TidyVerse
summary_stats <- cancer_data %>%
  summarize(
      min = min(survival, na.rm = TRUE),
      max = max(survival, na.rm = TRUE),
      mean = mean(survival, na.rm = TRUE),
      median = median(survival, na.rm = TRUE),
      sd = sd(survival, na.rm = TRUE),
      variance = var(survival, na.rm = TRUE)    
  )
print(summary_stats)
```

```r
# Identify all of the unique values in a column 
unique_values <- unique(cancer_data$organ)
print(unique_values)
```

```r
# Produces dataframe of numerical results using ggformula
cancer_data %>% df_stats(~survival, mean)
cancer_data %>% df_stats(~survival, sd)
cancer_data %>% df_stats(~survival, var)
cancer_data %>% df_stats(~survival, summary)
```

## Plot Data

```r
# Assuming cancer_reccurence_data is your data frame
# Create a dot plot using ggplot2
ggplot(cancer_reccurence_data, aes(x = tummor)) +
  geom_dotplot(binwidth = 2) +
  labs(title = "Recurrence of Cancer",
       x = "Tummor",
       y = "Count") +
  ylim(0, 1) + 
  theme_minimal()
```

```r
# ggformula plotting example
cancer_reccurence_data %>% gf_histogram(~tummor, title="Recurrence of Cancer")
```
