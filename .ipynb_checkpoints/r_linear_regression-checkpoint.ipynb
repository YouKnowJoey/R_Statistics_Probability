### Problem 1

11.13) A study of the amount of rainfall and the quantity of air pollution removed produced the following data: 
 
DailyRainfall, ParticulateRemoved,
 
 x(0.01cm) y(μg/m^3)
 
 4.3 126
 
 
 4.5 121
 
 5.9 116
 
 5.6 118
 
 6.1 114
 
 5.2 118
 
 3.8 132
 
 2.1 141
 
 7.5 108

 (a) Find the equation of the regression line to predict the particulate removed from the amount of daily rainfall.
 
 (b) Estimate the amount of particulate removed when the daily rainfall is x = 4.8 units.


```r
# Creating lists for the data
daily_rainfall <- c(4.3, 4.5, 5.9, 5.6, 6.1, 5.2, 3.8, 2.1, 7.5)
particulate_removed <- c(126, 121, 116, 118, 114, 118, 132, 141, 108)

# Performing linear regression
model <- lm(particulate_removed ~ daily_rainfall)

# Getting the coefficients of the regression line
coefficients <- coef(model)

# Printing the equation of the regression line
cat("Equation of the regression line: y =", coefficients[1], "+", coefficients[2], "* x\n")

# Estimating the amount of particulate removed for x = 4.8
x <- 4.8
predicted_particulate <- coefficients[1] + coefficients[2] * x
cat("Estimated particulate removed for daily rainfall of 4.8 units:", predicted_particulate, "μg/m³\n")
```

```r
# Creating lists for the data
daily_rainfall <- c(4.3, 4.5, 5.9, 5.6, 6.1, 5.2, 3.8, 2.1, 7.5)
particulate_removed <- c(126, 121, 116, 118, 114, 118, 132, 141, 108)

# Calculate the means
mean_x <- mean(daily_rainfall)
mean_y <- mean(particulate_removed)

# Calculate the components needed for the slope (b1) and intercept (b0)
numerator <- sum((daily_rainfall - mean_x) * (particulate_removed - mean_y))
denominator <- sum((daily_rainfall - mean_x)^2)

# Calculate the slope (b1) and intercept (b0)
b1 <- numerator / denominator
b0 <- mean_y - b1 * mean_x

# Printing the equation of the regression line
cat("Equation of the regression line: y =", b0, "+", b1, "* x\n")

# Estimating the amount of particulate removed for x = 4.8
x <- 4.8
predicted_particulate <- b0 + b1 * x
cat("Estimated particulate removed for daily rainfall of 4.8 units:", predicted_particulate, "μg/m³\n")
```

### Problem 2

11.14) A professor in the School of Business in a university polled a dozen colleagues about the number of professional meetings they attended in the past five years (x) and the number of papers they submitted to refereed journals (y) during the same period. The summary data are given as follows:



Fit a simple linear regression model between x and y by finding out the estimates of intercept and slope. Comment on whether attending more professional meetings would result in publishing more papers.

```r
# Given summary data
n <- 12
mean_x <- 4 # number of professional meetings attended in past five years
mean_y <- 12 # number of papers submitted to refereed journals 
sum_x_squared <- 232
sum_xy <- 318

# Calculate the slope (b1)
b1 <- (sum_xy - n * mean_x * mean_y) / (sum_x_squared - n * mean_x^2)

# Calculate the intercept (b0)
b0 <- mean_y - b1 * mean_x

# Print the slope and intercept
cat("Slope (b1):", b1, "\n")
cat("Intercept (b0):", b0, "\n")

# Interpretation
if (b1 > 0) {
  cat("Attending more professional meetings is associated with publishing more papers.\n")
} else if (b1 < 0) {
  cat("Attending more professional meetings is associated with publishing fewer papers.\n")
} else {
  cat("There is no association between attending meetings and publishing papers.\n")
}
```

### Problem 3

11.42) For a particular variety of plant, researchers wanted to develop a formula for predicting the quantity of seeds (in grams) as a function of the density of plants. They conducted a study with four levels of the factor x, the number of plants per plot. Four replications were used for each level of x. The data are shown as follows:



Is a simple linear regression model adequate for analyzing this data set?

```r
# Plants per plot
plants_per_plot <- c(10, 20, 30, 40)

# Quantity of seeds (as a single vector for all sets)
quantity_of_seeds <- c(12.6, 15.3, 17.9, 19.2, 
                       11.0, 16.1, 18.3, 19.6,
                       12.1, 14.9, 18.6, 18.9,
                       10.9, 15.6, 17.8, 20.0)

# Repeat plants per plot for each observation
plants_expanded <- rep(plants_per_plot, each = 4)

# Fit a simple linear regression model
model <- lm(quantity_of_seeds ~ plants_expanded)

# Summary of the model
summary(model)

# Plotting
plot(plants_expanded, quantity_of_seeds, main = "Scatter Plot with Regression Line",
     xlab = "Plants per Plot (grams)", ylab = "Quantity of Seeds",
     pch = 19)
abline(model, col = "red")

# Residuals plot
plot(resid(model), main = "Residuals Plot", ylab = "Residuals", xlab = "Fitted Values")
abline(h = 0, col = "red", lty = 2)
```

### Problem 5

12.26) For Exercise 12.8 on page 451, construct a 90% confidence interval for the mean compressive strength when the concentration is x = 19.5 and a quadratic model is used.

> 12.8 The following is a set of coded experimental data on the compressive strength of a
> particular alloy at various values of the concentration of some additive:
> 


```r
# Load necessary library
library(ggplot2)

# Define the data
concentration <- rep(c(10.0, 15.0, 20.0, 25.0, 30.0), each = 3)
strength <- c(25.2, 27.3, 28.7, 29.8, 31.1, 27.8, 31.2, 32.6, 29.7, 31.7, 30.1, 32.3, 29.4, 30.8, 32.8)

# Create a data frame
df <- data.frame(Concentration = concentration, Strength = strength)
print(df)

# Fit a quadratic model
quad_model <- lm(Strength ~ poly(Concentration, 2, raw = TRUE), data = df)

# Predict the mean compressive strength at x = 19.5 for the confidence interval at 90%
new_data <- data.frame(Concentration = 19.5)
pred <- predict(quad_model, newdata = new_data, interval = "confidence", level = 0.90)

# Print the prediction and confidence interval
print(pred)
```

## Analysis of Variance (ANOVA)

**Scenario**

>Suppose a researcher wants to determine if three different teaching methods (A, B, and C) have different effects on students' test scores. The researcher collects data from a random sample of students who were taught using one of the three methods and records their test scores.
>
>Hypothetical Data:
>
>We'll generate some test scores for each teaching method and perform a one-way ANOVA to see if there are any statistically significant differences in the means of the test scores across the different methods.

**Interpretation**

>Null Hypothesis (H0): There is no difference in the mean test scores between the different teaching methods.
>
>Alternative Hypothesis (H1): At least one teaching method leads to different mean test scores compared to the others.
>
>The output will show an F-statistic and a p-value. If the p-value is less than the significance level (commonly 0.05), we reject the null hypothesis, indicating that at least one method differs significantly in terms of mean test scores.

```r
# Set seed for reproducibility
set.seed(123)

# Generate hypothetical data
method_A <- rnorm(30, mean = 75, sd = 10)
method_B <- rnorm(30, mean = 80, sd = 10)
method_C <- rnorm(30, mean = 85, sd = 10)

# Combine data into a data frame
test_scores <- data.frame(
  scores = c(method_A, method_B, method_C),
  method = factor(rep(c("A", "B", "C"), each = 30))
)

# Perform ANOVA
anova_result <- aov(scores ~ method, data = test_scores)

# Display ANOVA summary
summary(anova_result)
```


Degrees of Freedom (Df):

    method has 2 degrees of freedom (since there are three groups, df = 3 - 1).
    Residuals (error) has 87 degrees of freedom (total sample size minus the number of groups, 90 - 3).
Sum of Squares (Sum Sq):

    method: The sum of squares between the groups.
    Residuals: The sum of squares within the groups (error).
    Mean Squares (Mean Sq):

Calculated by dividing the sum of squares by the respective degrees of freedom.
F Value:

    The F-statistic is the ratio of the mean square between the groups to the mean square within the groups.
Pr(>F):

    The p-value associated with the F-statistic. A low p-value (typically < 0.05) indicates that there is a statistically significant difference in means between at least one pair of groups.
Conclusion

    If the p-value is less than 0.05, you would reject the null hypothesis, concluding that there is a significant difference in the mean test scores among the teaching methods. This suggests that at least one teaching method is different in effectiveness compared to the others.
